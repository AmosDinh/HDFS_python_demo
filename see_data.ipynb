```python
from pathlib import Path
import os 

datapath = Path('data')/'data'
```

```python
# get the total filesize of all data in the data folder

for folder in os.listdir(datapath):
    sizes = 0
    for file in os.listdir(datapath/folder):
        sizes += os.path.getsize(datapath/folder/file)

    print(f'{folder}: {sizes / (1024 * 1024)} mb')
```
